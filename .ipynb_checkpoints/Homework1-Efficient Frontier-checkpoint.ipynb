### Import modules

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Import and read file_path "Industry_Portfolios.csv". Rename as per download file. Save the file in the same path as where this jupyter file is save

```python
file_path = "Industry_Portfolios.csv"
```

```python
data_frame = pd.read_csv(file_path)
```

### Show data_frame of "Industry Portfolio". This dataframe indicate returns for each portfolio from 2004 to 2013

```python
data_frame
```

### Drop "Date" column from the data_frame. Only remains return for each portfolio

```python
selected_columns = data_frame.drop('Date', axis=1)
selected_columns
```

```python
ten_industry_portfolios = selected_columns.columns.to_numpy()
ten_industry_portfolios
```

### Calculate the average return (mean) of 10 industry portfolio. Translate into R(10x1) and R_transposed(1x10)

```python
return_averages = selected_columns.mean()
R_transposed = return_averages.values
R = R_transposed.reshape(-1,1)
R
```

### Calculate the standard deviation ($\sigma$) of 10 industry portfolio

```python
std_dev_industry = selected_columns.std()
std_dev_industry_array = std_dev_industry.values
std_dev_industry_array
```

### Create covariance matrix dataframe

```python
covariance_matrix_df = selected_columns.cov()
covariance_matrix_df
```

### Calculate Alpha, Zeta, Delta, RMV using formula provided in slide 11

```python
covariance_matrix = covariance_matrix_df.values
inverse_covariance_matrix = np.linalg.inv(covariance_matrix)
e = np.ones((10,1))
e_transposed = e.transpose()
alpha = np.dot(np.dot(R_transposed, inverse_covariance_matrix),e).flatten()
zeta = np.dot(np.dot(R_transposed, inverse_covariance_matrix),R).flatten()
delta = np.dot(np.dot(e_transposed, inverse_covariance_matrix),e).flatten()
RMV = alpha/delta

# Display the calculated values
print(f"Alpha: {alpha}")
print(f"Zeta: {zeta}")
print(f"Delta: {delta}")
print(f"RMV: {RMV}")
```

### Create a range of values for Rp from 0 to 2 with an interval of 0.1

```python
Rp_range = np.arange(0, 2.1, 0.1).reshape(1, -1)
Rp_range
```

### Calculate the variance and standard deviation for each value of Rp using the formula provided in slide 13

```python
variance = (1 / delta + (delta / (zeta * delta - alpha**2)) * (Rp_range - RMV)**2).flatten()
std_dev = np.sqrt(variance)
```

```python
print(f"variance: {variance}\n")
print(f"std deviation: {std_dev}")
```

### Create a table showing the mean return and standard deviation of return for the ten industry portfolios.

```python
data = pd.DataFrame({
    'Industry_Portfolios': ten_industry_portfolios,
    'Mean Return': R_transposed,
    'Standard Deviation': std_dev_industry_array
}, index=range(1, 11))

# Use the `style` attribute to align text at the center
styled_data = data.style.set_properties(**{'text-align': 'center'})

styled_data
```

### Calculate the Sharpe ratio (slide 30) for the tangency portfolio, and also the tangency portfolio weights (slide 29) for the ten industry portfolios.

### In this code. Risk free rate is coded as 0.13%. Change if it differs

```python
# Calculate CAL Line for Asset Allocation with Riskless Asset

# In the homework it is indicated that Rf = 0.13%
Rf = 0.13
last_index_Rp = Rp_range[-1]
last_index_std_dev = std_dev[-1]
CAL_x_axis = np.array([0, last_index_std_dev])
sharpe_ratio = ((zeta- 2*alpha*Rf + delta * Rf**2)**(1/2)).item()

# Calculate the tangency portfolio standard deviation
std_dev_tg = -(zeta-2*alpha*Rf+delta*Rf**2)**(1/2)/(delta*(Rf-RMV))
Rtg = sharpe_ratio * std_dev_tg + Rf

# Calculate tangency portfolio weight
lagrange_multiplier = ((Rtg-Rf)/(zeta-2*alpha*Rf+delta*Rf**2)).flatten()
lagrange_multiplier
Rf_matrix = Rf*e
portfolio_weight = lagrange_multiplier * np.dot(inverse_covariance_matrix , (R-Rf_matrix))


print(f"Sharpe Ratio: {sharpe_ratio}")
print(f"Sigma_tg: {std_dev_tg}")
print(f"Return_tg: {Rtg}")
print(f"Tangency Portfolio Weight: {portfolio_weight}")
```

### Summarize data frame show mean, std_dev, tangency portfolio weight

```python
data['Porftolio Weight'] = portfolio_weight
data
```

### Plot the minimum-variance frontier (without the riskless asset) generated by the ten industry portfolios & efficient frontier (with the riskless asset) on the same graph with risk-free rate 0.13%

```python
# Flatten Rp_range and std_dev to 1D arrays
Rp_range = Rp_range.flatten()
std_dev = std_dev.flatten()

# plt.scatter(std_dev[upper_half_indices], Rp_range[upper_half_indices], marker='o', color='blue')
plt.figure(figsize=(10, 6))
plt.plot(std_dev,
         Rp_range,
         color='blue',
         linestyle='-',
         label='Without riskless asset')

# Draw a horizontal line at RMV
plt.axhline(y=RMV, color='green', linestyle='dotted')

# Find the index where the Sharpe ratio is maximized (tangent point)
tangent_idx = np.argmax(sharpe_ratio)
tangent_std_dev = std_dev[tangent_idx]
tangent_return = Rp_range[tangent_idx]

# Plot a linear line between using the Sharpe ratio as the slope
linear_line = Rf + sharpe_ratio * CAL_x_axis
plt.plot(CAL_x_axis, linear_line, color='red', linestyle='-', label='With riskless asset')

# Plot the CAL tangent point
plt.scatter(std_dev_tg, Rtg, color='blue', marker='o')
plt.annotate(f'Tangent \n{round(std_dev_tg.item(),2),round(Rtg.item(),2)}'
             , (std_dev_tg-0.4, Rtg)
             , ha='center'
             , fontsize=8
             , weight='bold')

# Set labels and title
plt.xlabel('Std Dev of Return (%)')
plt.ylabel('Mean Return (%)')
# plt.title('Efficient Frontier',
#           weight='bold')

# Set the Y-axis ticks with an increment of 0.1
plt.yticks(np.arange(0, max(Rp_range) + 0.1, 0.1))

# Set the X-axis ticks with an increment of 0.1
plt.xticks(np.arange(0, max(std_dev) + 0.1, 0.5))

# Label the y-axis at the RMV intersection point
plt.text(0, RMV, 'RMV', ha='left', va='bottom', color='blue')
plt.text(-0.2, Rf, f'Rf={Rf}', ha='right', va='bottom', color='blue')

# Set the X-axis limits to start from 0
plt.xlim(0, max(std_dev)+0.5)  # Adjust the upper limit as needed
plt.ylim(0, max(Rp_range))  # Adjust the upper limit as needed
plt.title("Portfolio Frontiers", fontsize=12, fontweight='bold')

# Show the plot
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2)  # Adjust the position and number of columns
plt.grid(True, linestyle='dotted')
plt.show()
```

# Economic significance:

Minimum-variance frontier represents outermost envelope of attainable portfolios â‡’ no attainable portfolio exists on the left of the minimum-variance frontier.

Efficient frontier consists of portfolios with highest potential reward for given amount of risk â‡’ risk-averse investor must invest in (optimal) portfolio on efficient frontier to maximise expected utility.

Tangency portfolio has the highest possible Sharpe ratio.
